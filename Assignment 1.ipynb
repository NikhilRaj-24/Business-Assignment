```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

df = pd.read_csv('C:/Users/nikhi/Visual Codes/Business Analytics/shopping_trends.csv')
df
```

```python
#Q1
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=10, edgecolor='black', color='skyblue')
plt.title('Age Distribution of Shoppers')
plt.xlabel('Age Group')
plt.ylabel('Number of Shoppers')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()

most_shopped_age_group = df['Age'].value_counts().idxmax()
count_most_shopped_age_group = df['Age'].value_counts().max()

# Print the result
print(f"The age group that has shopped the most is: {most_shopped_age_group} years old. They have shopped {count_most_shopped_age_group} times.")

# Calculate the top 3 most shopped age groups
top_3_most_shopped_age_groups = df['Age'].value_counts().nlargest(3)

# Print the top 3 most shopped age groups
print("Top 3 Most Shopped Age Groups:")
for age_group, count in top_3_most_shopped_age_groups.items():
    print(f"Age Group: {age_group} years old | Number of Shoppers: {count}")
```

Here we can see old people which are around 40's, 60's and 70's have shopped the most.

```python
#Q2
average_age = df['Age'].mean()
print(f"The average age of the customers is: {average_age:.2f} years old.")
```

We can understand that people around 40's who are parents or finished their 30's are shopping the most.

```python
#Q3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Get the most commonly purchased item
most_common_item = df['Item Purchased'].mode().iloc[0]

# Plot the most commonly purchased item
plt.figure(figsize=(12, 6))
sns.countplot(x='Item Purchased', data=df, order=df['Item Purchased'].value_counts().index)
plt.title('Most Commonly Purchased Item')
plt.xlabel('Item')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Here we can understand most of the items here are Dresses such as Blouses, Pants, Shirts, Jackets, etc.

```python
most_common_item = df['Item Purchased'].mode().values[0]
print(f"The most commonly purchased item is: {most_common_item}.")
```

```python
#Q4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the total purchase amount for each category
total_purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].sum().reset_index()

# Plot the total purchase amount for each category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Purchase Amount (USD)', data=total_purchase_by_category)
plt.title('Total Purchase Amount for Each Category')
plt.xlabel('Category')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45, ha='right')

# Annotate the total purchase amount on the bars
for index, value in enumerate(total_purchase_by_category['Purchase Amount (USD)']):
    plt.text(index, value + 50, f'${value:.2f}', ha='center', va='bottom', fontsize=8)

plt.show()
```

As we saw in the earlier question, most of the purchases are related to clothing. So, we can say that most of the people are buying clothing items.

```python
total_purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].sum()
print("Total purchase amount for each category:")
print(total_purchase_by_category)
```

The Clothing category has been purchased by the customers the most by 104264, followed by Accessories - 74200, Footwear - 36093, Outwear - 18524.

```python
#Q5
average_rating_male = df[df['Gender'] == 'Male']['Review Rating'].mean()
average_rating_female = df[df['Gender'] == 'Female']['Review Rating'].mean()
print(f"The average review rating for male customers is: {average_rating_male:.2f}")
print(f"The average review rating for female customers is: {average_rating_female:.2f}")
```

```python
#Q6
min_purchase_amount = df['Purchase Amount (USD)'].min()
median_purchase_amount = df['Purchase Amount (USD)'].median()
mean_purchase_amount = df['Purchase Amount (USD)'].mean()
max_purchase_amount = df['Purchase Amount (USD)'].max()

print(f"Minimum Purchase Amount: {min_purchase_amount:.2f}")
print(f"Median Purchase Amount: {median_purchase_amount:.2f}")
print(f"Mean Purchase Amount: {mean_purchase_amount:.2f}")
print(f"Maximum Purchase Amount: {max_purchase_amount:.2f}")
```

Here we can see that the median and mean close to each other, suggesting a somewhat symmetrical distribution. The presence of a maximum value of $100.00 indicates a range of transaction sizes, with the majority falling between $20.00 and $60.00.

```python
#Q7
promo_code_users_count = df[df['Promo Code Used'] == 'Yes'].shape[0]
non_promo_code_users_count = df[df['Promo Code Used'] == 'No'].shape[0]

#plot it
plt.figure(figsize=(8, 6))
sns.countplot(x='Promo Code Used', data=df)
plt.title('Number of Customers Who Used a Promo Code')
plt.xlabel('Promo Code Used')
plt.ylabel('Number of Customers')
plt.show()

print(f"The number of customers who used a promo code is: {promo_code_users_count}")
print(f"The number of customers who did not use a promo code is: {non_promo_code_users_count}")
```

We can see that the customers have used 1677 time promo codes which is really a high number. People have bought items without promo codes for 2223 times. So, we can say that people are using promo codes to buy items.

```python
#Q10
payment_method_counts = df['Preferred Payment Method'].value_counts()

plt.figure(figsize=(10, 6))
payment_method_counts.plot(kind='bar', color='skyblue')
plt.title('Preferred Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Q11
import seaborn as sns

subscription_proportion = df['Subscription Status'].value_counts(normalize=True)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
subscription_proportion.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Subscription Proportion')

plt.subplot(1, 2, 2)
sns.scatterplot(x='Frequency of Purchases', y='Subscription Status', data=df, hue='Subscription Status', palette=['lightcoral', 'skyblue'])
plt.title('Purchase Frequency vs Subscription')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Subscription Status')

plt.tight_layout()

plt.show()
```

The plot indicates a lower proportion of subscribed customers compared to non-subscribed ones.

```python
#Q12
purchase_by_location = df.groupby('Location')['Purchase Amount (USD)'].sum()

# Sorting and getting the top 10 locations
top_10_locations = purchase_by_location.sort_values(ascending=False).head(10)

# Plotting the bar chart
plt.figure(figsize=(12, 6))
top_10_locations.plot(kind='bar', color='skyblue')
plt.title('Top 10 Locations Based on Purchase Amount')
plt.xlabel('Location')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

Montana is where customers have spent the most followed by Illinois, Calofornia, Idaho, Nevada, Alabama, New York, North Dakota, Westr Virginia, Nebraska.
